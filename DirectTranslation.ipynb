```python
len(es_deleted_index)
```

```python
es_labels = []
i = 0
j = 0
with open('en_train/english_train.labels') as f:
    for line in f:
        if j < len(es_deleted_index) and i == es_deleted_index[j]:
            j+=1
            i+=1
            continue
        es_labels.append(int(line.strip()))
        i+=1
with open('test/english_test.labels') as f:
    for line in f:
        if j < len(es_deleted_index) and i == es_deleted_index[j]:
            j+=1
            i+=1
            continue
        es_labels.append(int(line.strip()))
        i+=1
```

```python
en_labels = []
i = 0
j = 0
with open('es_train/spanish_train.labels') as f:
    for line in f:
        if j < len(en_deleted_index) and i == en_deleted_index[j]:
            j+=1
            i+=1
            continue
        en_labels.append(int(line.strip()))
        i+=1
with open('test/spanish_test.labels') as f:
    for line in f:
        if j < len(en_deleted_index) and i == en_deleted_index[j]:
            j+=1
            i+=1
            continue
        en_labels.append(int(line.strip()))
        i+=1
```

```python
es_out = []
es_labels_out = []
for i in range(len(es_labels)):
    if es_labels[i] in [4, 10 ,12 ,15, 17, 18]:
        continue
    if es_labels[i] == 5:
        es_labels_out.append(4)
    elif es_labels[i] == 6:
        es_labels_out.append(10)
    elif es_labels[i] == 7:
        es_labels_out.append(15)
    elif es_labels[i] == 8:
        es_labels_out.append(11)
    elif es_labels[i] == 9:
        es_labels_out.append(5)
    elif es_labels[i] == 11:
        es_labels_out.append(9)
    elif es_labels[i] == 19:
        es_labels_out.append(13)
    else :es_labels_out.append(es_labels[i])
    es_out.append(es_trains[i])  
```

```python
en_out = []
en_labels_out = []
for i in range(len(en_labels)):
    if en_labels[i] in [4, 10 ,12 ,15, 17, 18]:
        continue
    if en_labels[i] == 5:
        en_labels_out.append(4)
    elif en_labels[i] == 6:
        en_labels_out.append(10)
    elif en_labels[i] == 7:
        en_labels_out.append(15)
    elif en_labels[i] == 8:
        en_labels_out.append(11)
    elif en_labels[i] == 9:
        en_labels_out.append(5)
    elif en_labels[i] == 11:
        en_labels_out.append(9)
    elif en_labels[i] == 19:
        en_labels_out.append(13)
    else :en_labels_out.append(en_labels[i])
    en_out.append(en_trains[i])  
```
